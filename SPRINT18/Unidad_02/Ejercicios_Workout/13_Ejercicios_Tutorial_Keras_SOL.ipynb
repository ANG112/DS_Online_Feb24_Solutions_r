### Ejercicio 7: Construye el modelo con Keras

Construyamos nuestro modelo. Aqu, utilizaremos un modelo `secuencial` con dos capas ocultas densamente conectadas y una capa de salida que devuelve un nico valor continuo. Es decir una MLP con dos capas ocultas.

Por tanto, tendr tres capas. Una inicial con activacin relu (usa el argumento `input_shape` para darle el nmero de features). Otra hidden layer con activacin relu y la de salida de regresin, que se compondr de una nica neurona. Pon las neuronas que consideres interesantes para las dos primeras capas (NOTA: En este caso partimos de pocas features, por lo que puede ser interesante poner 2 o 3 como mnimo el nmero de features para poder obtener unas pseudofeatures que intenten capturar las interrelaciones.)

Otros datos:

Para el compile utiliza un loss='mse'.

optimizer = `tf.keras.optimizers.RMSprop(0.001)`

Y en metrics aade en una lista el `mae` y `mse`.

Utiliza una aproximacin en prisma (mismo numero de unidades en las capas ocultas)

````python
factor_expansion = 3
num_units = len(features) * factor_expansion
factor_expansion
````

*Mtodo 1, como aadido de capas*

````python
model_add = keras.models.Sequential()

model_add.add(keras.layers.Dense(units = num_units, activation='relu', input_shape = [len(features)]))
model_add.add(keras.layers.Dense(units = num_units, activation='relu'))
model_add.add(keras.layers.Dense(units = 1))
````

*Mtodo 2, usando las capas en los argumentos de la inicializacin*

````python

model = keras.Sequential([
layers.Dense(num_units, activation='relu', input_shape=[len(features)]),
layers.Dense(num_units, activation='relu'),
layers.Dense(1)
])


model.compile(loss='mse',
              optimizer=tf.keras.optimizers.RMSprop(0.001),
              metrics=['mae', 'mse'])
````

### Ejercicio 10

Visualiza el progreso de entrenamiento del modelo usando las estadsticas almacenadas en el objeto `history`. Muestra la evolucin del error en train y validation por un lado, y la evolucin de la prdida (loss), tambin train contra validation, en otra.

````python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
````

````python
def plot_history(history, error = True):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  if error:
    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Abs Error [MPG]')
    plt.plot(hist['epoch'], hist['mae'],
            label='Train Error')
    plt.plot(hist['epoch'], hist['val_mae'],
            label = 'Val Error')
    plt.ylim([0,5])
    plt.legend()

    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Square Error [$MPG^2$]')
    plt.plot(hist['epoch'], hist['mse'],
            label='Train Error')
    plt.plot(hist['epoch'], hist['val_mse'],
            label = 'Val Error')
    plt.ylim([0,20])
    plt.legend()
    plt.show()

  else:
    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.plot(hist['epoch'], hist['loss'],
            label='Train loss')
    plt.plot(hist['epoch'], hist['val_loss'],
            label = 'Val loss')
    plt.legend()
    plt.show()



plot_history(history)
````

### Ejercicio 12

Independientemente de lo comentado en 11, vamos a aadir un callback al entrenamiento. Para ello, construye el modelo otra vez y vuelve a compilarlo (si no lo haces, como ya sabes comenzar el entrenamiento donde lo dej y no nos valdr el callback para nada). Entrena el modelo con los mismos hiperparmetros del ejercicio 10 y pinta la grfica de evolucin de los errores e interprtalo (obten el MAE medio una vez "estabilizado" el entrenamiento).  

Para el earlystopping utiliza un margen de 20 epochs.

````python
model = keras.Sequential([
layers.Dense(num_units, activation='relu', input_shape=[len(features)]),
layers.Dense(num_units, activation='relu'),
layers.Dense(1)
])


model.compile(loss='mse',
            optimizer=tf.keras.optimizers.RMSprop(0.001),
            metrics=['mae', 'mse'])


# The patience parameter is the amount of epochs to check for improvement
history = model.fit(X_train,
                    y_train, epochs=1000,
                    validation_split = 0.2,
                    verbose= 1,
                    callbacks=[keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)])

plot_history(history)
````

Interpreta el grfico

*Est claro que no mereca la pena entrenar con tantas pocas, antes de las 200 hemos terminado y que sobre las 80 epocas se estabiliza*

````python
hist = pd.DataFrame(history.history)
hist[80:].mae.hist()
````

````python
mae = hist[80:].mae.mean()
mae
````

*El grfico muestra que en el conjunto de validacin, el error promedio generalmente es de alrededor de +/- 2 MPG. Es esto bueno? Le dejaremos la decisin a "negocio"*
